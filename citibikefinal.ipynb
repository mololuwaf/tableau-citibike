```python
import pandas as pd
```

```python
#2013 CSV's
july_13 = pd.read_csv('2013-07 - Citi Bike trip data.csv')
august_13 = pd.read_csv('2013-08 - Citi Bike trip data.csv')
september_13 = pd.read_csv('2013-09 - Citi Bike trip data.csv')
october_13 = pd.read_csv('2013-10 - Citi Bike trip data.csv')
november_13 = pd.read_csv('2013-11 - Citi Bike trip data.csv')
december_13 = pd.read_csv('2013-12 - Citi Bike trip data.csv')
```

```python
bike_data13 = pd.concat([july_13, august_13, september_13, october_13, november_13, december_13]).reset_index(drop=True)
print(bike_data13.shape)
print(bike_data13.columns)
bike_data13.sample(5)
```

```python
# 2014 csvs
january_14 = pd.read_csv('2014-01 - Citi Bike trip data.csv')
february_14  = pd.read_csv('2014-02 - Citi Bike trip data.csv')
march_14  = pd.read_csv('2014-03 - Citi Bike trip data.csv')
april_14  = pd.read_csv('2014-04 - Citi Bike trip data.csv')
may_14  = pd.read_csv('2014-05 - Citi Bike trip data.csv')
june_14  = pd.read_csv('2014-06 - Citi Bike trip data.csv')
july_14  = pd.read_csv('2014-07 - Citi Bike trip data.csv')
august_14  = pd.read_csv('2014-08 - Citi Bike trip data.csv')
september_14  = pd.read_csv('201409-citibike-tripdata.csv')
october_14  = pd.read_csv('201410-citibike-tripdata.csv')
november_14  = pd.read_csv('201411-citibike-tripdata.csv')
december_14  = pd.read_csv('201412-citibike-tripdata.csv')
```

```python
bike_data14 = pd.concat([january_14, february_14, march_14, april_14, may_14, june_14, july_14, august_14, september_14, october_14, november_14, december_14]).reset_index(drop=True)
print(bike_data14.shape)
print(bike_data14.columns)
bike_data14.sample(5)
```

```python
# Save the dataframe to a CSV file
bike_data14.to_csv('citibikedatafinal14.csv', index=False)
```
